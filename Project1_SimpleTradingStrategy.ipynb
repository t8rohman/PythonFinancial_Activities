# Creating Features and Columns in Data Frame

Now, I want to add new column by calculating existing columns that needed to build a simple stock strategy.

```python
# Create Price Difference column
# Formula = (Close Price at d+1) - (Close Price at d)

df_fb['PriceDiff'] = df_fb['Close'].shift(-1) - df_fb['Close']
df_ms['PriceDiff'] = df_ms['Close'].shift(-1) - df_fb['Close']

print(df_fb.head(3))
print(df_ms.head(3))
```

```python
# Create Daily Return column
# Formula = Price Difference / Close Price

df_fb['DailyReturn'] = df_fb['PriceDiff'] / df_fb['Close']
df_ms['DailyReturn'] = df_ms['PriceDiff'] / df_ms['Close']

print(df_fb.head(3))
print(df_ms.head(3))
```

```python
# Create Direction column
# +1 if profit, 0 if same or loss

df_fb['PL'] = [1 if df_fb['PriceDiff'].loc[ei] > 0 else 0 for ei in df_fb.index]
df_ms['PL'] = [1 if df_ms['PriceDiff'].loc[ei] > 0 else 0 for ei in df_ms.index]

print(df_fb[['PriceDiff', 'PL']].head(5))
print(df_ms[['PriceDiff', 'PL']].head(5))
```

```python
# Create Moving Average (MA) Column
# Moving Average created by calculating closing price mean in the last n days
# For example, MA40 means that mean in the last 40 days
# Using rolling function in pandas data frame

df_fb['MA10'] = df_fb['Close'].rolling(10).mean()
df_fb['MA50'] = df_fb['Close'].rolling(50).mean()

df_ms['MA10'] = df_ms['Close'].rolling(10).mean()
df_ms['MA50'] = df_ms['Close'].rolling(50).mean()

print(df_fb[['Close', 'MA10', 'MA50']].tail(5))
print(df_ms[['Close', 'MA10', 'MA50']].tail(5))
```

# Stock Decision Strategy

Let's now plot and make the calculation for the decision strategy. Using MA10 and MA50, we can make a simple decision whether to long on a stock or to hold making a decision.

```python
# Ploting the MA10 and MA50 for the strategy
# Facebook stock in 2016

plt.figure(figsize = (10, 8))
df_fb['Close'].loc['2016-01-01' : '2016-12-31'].plot(color = 'gray')
df_fb['MA10'].loc['2016-01-01' : '2016-12-31'].plot(color = 'red')
df_fb['MA50'].loc['2016-01-01' : '2016-12-31'].plot(label = 'blue')
plt.show()
```

```python
# Ploting the MA10 and MA50 for the strategy
# Facebook stock in 2016

plt.figure(figsize = (10, 8))
df_ms['Close'].loc['2016-01-01' : '2016-12-31'].plot(color = 'gray')
df_ms['MA10'].loc['2016-01-01' : '2016-12-31'].plot(color = 'red')
df_ms['MA50'].loc['2016-01-01' : '2016-12-31'].plot(label = 'blue')
plt.show()
```

```python
# Creating strategy matrix
# MA10 is considered as fast signal, while MA50 is slow signal
# In this strategy, if MA10 > MA50, I am going to buy stock. If else, no.

df_fb['Decision'] = [1 if df_fb['MA10'].loc[ei] > df_fb['MA50'].loc[ei] else 0
                    for ei in df_fb.index]
df_ms['Decision'] = [1 if df_ms['MA10'].loc[ei] > df_ms['MA50'].loc[ei] else 0
                    for ei in df_ms.index]

print(df_fb[['MA10', 'MA50', 'Direction']].tail())
print(df_ms[['MA10', 'MA50', 'Direction']].tail())
```

```python
# Create daily profit sum

df_fb['Close1'] = df_fb['Close'].shift(-1)
df_fb['Profit'] = [df_fb.loc[ei, 'Close1'] - df_fb.loc[ei, 'Close'] if df_fb.loc[ei, 'Decision'] == 1 else 0
                  for ei in df_fb.index]

df_ms['Close1'] = df_ms['Close'].shift(-1)
df_ms['Profit'] = [df_ms.loc[ei, 'Close1'] - df_ms.loc[ei, 'Close'] if df_ms.loc[ei, 'Decision'] == 1 else 0
                  for ei in df_ms.index]

print(df_fb[['Decision', 'MA10', 'MA50', 'Close', 'Close1', 'Profit']].tail(5))
print(df_ms[['Decision', 'MA10', 'MA50', 'Close', 'Close1', 'Profit']].tail(5))
```

```python
# Plot the daily profit in y = 0 axis 
# for Facebook stock

plt.figure(figsize = (10, 8))
df_fb['Profit'].plot()
plt.axhline(y = 0, color = 'red')
```

```python
# Plot the daily profit in y = 0 axis 
# for Microsoft stock

plt.figure(figsize = (10, 8))
df_ms['Profit'].plot()
plt.axhline(y = 0, color = 'red')
```

```python
# Calculate the cumulative Profit

df_fb['Total_Wealth'] = df_fb['Profit'].cumsum()
df_ms['Total_Wealth'] = df_ms['Profit'].cumsum()

# Final sum of the money

print("Facebook: " + str(df_fb.loc['2018-02-02', 'Total_Wealth']))
print("Microsoft: " + str(df_ms.loc['2018-02-02', 'Total_Wealth']))
```
